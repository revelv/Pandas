```python
import pandas as pd
import numpy as np

print(pd.__version__)
print(np.__version__)
```

```python
n_rows = 40
n_cols = 5
cols = tuple('ABCDE')

df = pd.DataFrame(np.random.randint(1, 20, size=(n_rows, n_cols)), 
                  columns=cols)
df.head()
```

```python
### Line Plot
```

```python
df.head()
```

```python
df.plot(kind='line')
```

```python
df[['A', 'B']].plot(kind='line')
```

```python
### Bar Plot
```

```python
df.head()
```

```python
df.plot(kind='bar')
```

```python
df[['A', 'B']].plot(kind='bar')
```
